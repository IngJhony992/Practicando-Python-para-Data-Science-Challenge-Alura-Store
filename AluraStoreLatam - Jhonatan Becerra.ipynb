### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
#Ingreso total por cada tienda
ingreso_total_tienda1 = tienda1['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()
print(f"El ingreso total de la tienda 1 es: {ingreso_total_tienda1}")
print(f"El ingreso total de la tienda 2 es: {ingreso_total_tienda2}")
print(f"El ingreso total de la tienda 3 es: {ingreso_total_tienda3}")
print(f"El ingreso total de la tienda 4 es: {ingreso_total_tienda4}")
plt.bar(['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'], [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4], color="#FF7043")
plt.ticklabel_format(style='plain', axis='y')
plt.title("Ingreso total por cada tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingreso Total")

```

# 2. Ventas por categoría

```python
#Ventas por categoría tienda 1
ventas_por_categoria_tienda1 = tienda1.groupby(['Categoría del Producto']).size().reset_index(name='CantidadVendida')
categorias_populares_tienda1 = ventas_por_categoria_tienda1.sort_values(by='CantidadVendida', ascending=False)
print("Las categorías más populares de la tienda 1 son: ")
print(categorias_populares_tienda1)
print(" ")

#Gráfica Tienda 1
colores = ["#ECB649","#66BB6A","#FF8A65","#BA68C8","#4DD0E1", "#FBF6A4", "#F48A94", "#874F3E"]
plt.pie(categorias_populares_tienda1['CantidadVendida'], labels=categorias_populares_tienda1['Categoría del Producto'], autopct="%0.1f %%",colors=colores)
plt.title('Productos más y menos vendidos por categoría en la tienda 1')
```

```python
#Ventas por categoría tienda 2
ventas_por_categoria_tienda2 = tienda2.groupby(['Categoría del Producto']).size().reset_index(name='CantidadVendida')
categorias_populares_tienda2 = ventas_por_categoria_tienda2.sort_values(by='CantidadVendida', ascending=False)
print("Las categorías más populares de la tienda 2 son: ")
print(categorias_populares_tienda2)
print(" ")

#Gráfica Tienda 2
plt.pie(categorias_populares_tienda2['CantidadVendida'], labels=categorias_populares_tienda2['Categoría del Producto'], autopct="%0.1f %%", colors=colores)
plt.title('Productos más y menos vendidos por categoría en la tienda 2')
```

```python
#Ventas por categoría tienda 3
ventas_por_categoria_tienda3 = tienda3.groupby(['Categoría del Producto']).size().reset_index(name='CantidadVendida')
categorias_populares_tienda3 = ventas_por_categoria_tienda3.sort_values(by='CantidadVendida', ascending=False)
print("Las categorías más populares de la tienda 3 son: ")
print(categorias_populares_tienda3)
print(" ")

#Gráfica Tienda 3
plt.pie(categorias_populares_tienda3['CantidadVendida'], labels=categorias_populares_tienda3['Categoría del Producto'], autopct="%0.1f %%", colors=colores)
plt.title('Productos más y menos vendidos por categoría en la tienda 3')
```

```python
#Ventas por categoría tienda 4
ventas_por_categoria_tienda4 = tienda4.groupby(['Categoría del Producto']).size().reset_index(name='CantidadVendida')
categorias_populares_tienda4 = ventas_por_categoria_tienda4.sort_values(by='CantidadVendida', ascending=False)
print("Las categorías más populares de la tienda 4 son: ")
print(categorias_populares_tienda4)
print(" ")

#Gráfica Tienda 4
plt.pie(categorias_populares_tienda4['CantidadVendida'], labels=categorias_populares_tienda4['Categoría del Producto'], autopct="%0.1f %%", colors=colores)
plt.title('Productos más y menos vendidos por categoría en la tienda 4')
```

```python
# Renombrar columnas para identificar tiendas
categorias_populares_tienda1 = categorias_populares_tienda1.rename(columns={'CantidadVendida': 'Tienda 1'})
categorias_populares_tienda2 = categorias_populares_tienda2.rename(columns={'CantidadVendida': 'Tienda 2'})
categorias_populares_tienda3 = categorias_populares_tienda3.rename(columns={'CantidadVendida': 'Tienda 3'})
categorias_populares_tienda4 = categorias_populares_tienda4.rename(columns={'CantidadVendida': 'Tienda 4'})

# Unir todas las tablas por categoría
comparativa = categorias_populares_tienda1.merge(
    categorias_populares_tienda2, on='Categoría del Producto', how='outer').merge(
    categorias_populares_tienda3, on='Categoría del Producto', how='outer').merge(
    categorias_populares_tienda4, on='Categoría del Producto', how='outer')

# Rellenar NaN con 0 por si alguna tienda no vendió alguna categoría
comparativa = comparativa.fillna(0)

# Establecer la columna Categoría como índice para la gráfica
comparativa = comparativa.set_index('Categoría del Producto')

# Gráfica de barras agrupadas
comparativa.plot(kind='bar', figsize=(12,6), color=["#ECB649","#66BB6A","#FF8A65","#BA68C8"])

plt.title("Comparación de Ventas por Categoría entre Tiendas")
plt.xlabel("Categoría del Producto")
plt.ylabel("Cantidad Vendida")
plt.legend(title="Tiendas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
valoraciones = [valoracion_media_tienda1, valoracion_media_tienda2, valoracion_media_tienda3, valoracion_media_tienda4]
plt.figure(figsize=(8,5))
plt.plot(tiendas, valoraciones, marker='o', linestyle='-', linewidth=2)

plt.title("Valoración media por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación promedio")
plt.grid(True)
```
